```python
import requests
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
r = requests.get('https://collectionapi.metmuseum.org/public/collection/v1/search?q=sunflowers')
res = r.json()
obj_id_list = res['objectIDs']

for i in obj_id_list:
    if i==436524:
        print(i)
    elif i==20149:
        print(i)
    elif i==16822570:
        print(i)
    elif i==2032:
        print(i)
    elif i==437329:
        print(i)    
```

```python
df = pd.DataFrame()
```

```python

for i in obj_id_list:
        r_test = requests.get(f'https://collectionapi.metmuseum.org/public/collection/v1/objects/'+str(i))
        obj_test = r_test.json()
        
        try:
            df = df.append({'objectID': obj_test["objectID"],
                       'title': obj_test["title"],
                       'artistDisplayName': obj_test["artistDisplayName"],
                       'department': obj_test["department"],
                       'objectBeginDate': obj_test["objectBeginDate"],
                       'objectEndDate': obj_test["objectEndDate"],
                       'period': obj_test["period"],
                       'objectName': obj_test["objectName"],
                       'culture': obj_test["culture"]}, ignore_index=True)
        except:    
            continue
df        
```

```python
df.query('culture == "Japan"')\
    .groupby('culture')\
    .agg({'objectID' : 'count'})
```

```python
r = requests.get('https://collectionapi.metmuseum.org/public/collection/v1/search?departmentId=6&q=cat')
cats = r.json()
cat_ids = cats['objectIDs']
new_cat_ids = cat_ids[:1000]
df_2 = pd.DataFrame()
for i in new_cat_ids:
        r_test = requests.get(f'https://collectionapi.metmuseum.org/public/collection/v1/objects/'+str(i))
        obj_test = r_test.json()
        
        try:
            df_2 = df_2.append({'objectID': obj_test["objectID"],
                       'title': obj_test["title"],
                       'artistDisplayName': obj_test["artistDisplayName"],
                       'department': obj_test["department"],
                       'objectBeginDate': obj_test["objectBeginDate"],
                       'objectEndDate': obj_test["objectEndDate"],
                       'period': obj_test["period"],
                       'objectName': obj_test["objectName"],
                       'culture': obj_test["culture"]}, ignore_index=True)
        except:    
            continue
df_2        
```

```python
df_2.groupby('period')\
    .agg({'objectID' : 'count'})\
    .sort_values('objectID', ascending=False)
```
