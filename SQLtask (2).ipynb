```python
#                                    Пример использования конструкции WITH  (для объявления переменных и параметров запроса)

-- объявляем параметры
WITH
            (SELECT AVG(InstallCost) FROM installs) AS avg_install_cost,
            sum_rub * 100 AS rub_cent
-- используем
SELECT
            UserID,
            sum_rub - avg_install_cost AS economy,
            rub_cent
FROM (
            SELECT
            UserID,
            SUM(Rub) AS sum_rub
            FROM
            checks
            GROUP BY
            UserID
)
LIMIT 100
```

```python
#                                     Оконные функции

SELECT
    city,
    date,
    revenue,
    SUM(revenue) OVER w AS revenue_sum 
    row_number() OVER w AS rnum        --- оконная функция(только для окна) порядковый номер строки
    LAG() OVER w AS  ...                ---оконная функция СМЕЩЕНИЕ - знач с предидущ стр
    LID() OVER w AS  ...                ---оконная функция СМЕЩЕНИЕ - знач сo след стр
    
FROM
    data
   
 WINDOW w AS
(
    PARTITION BY city
    ORDER BY date ASC ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW
)
ORDER BY
    city ASC,
    date ASC
    
#                           ... UNBOUNDED PRECEDING AND UNBOUNDED FOLOWING (PRECEDING,CURRENT, FOLOWING  )
#                            ... 1 PRECEDING AND 2 FOLOWING

# ROWS    - есть еще RANGE,(по строкам иногда нельзя тк певая стр 01.01.20, а вторая 01.02.20 - пропуск в месяц,хотя стр 2-я)
# RANGE BETWEEN '1 day' PRECEDING AND '2 days' FOLOWING

#  WINDOW w AS()   - без параметров вся таблица попадает в окно


# SELECT
#     city,
#     date,
#     revenue,
#     SUM(revenue) OVER () AS revenue_sum - объявление ОКНА без блока WINDOW( тут без параметров)
```
