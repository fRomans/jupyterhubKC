```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline 
```

```python
datapath = '/mnt/HC_Volume_18315164/home-jupyter/jupyter-rom-fadeev/shared/homeworks/python_ds_miniprojects/4/data/'
```

```python
list_data = os.listdir(datapath)
```

```python
list_data
```

```python
df_1 = pd.DataFrame(columns=['name', 'date'])
new_df = pd.DataFrame()

for data in list_data:
    data_name = datapath + data + '/'
    for namDN in os.listdir(data_name):
            full_path = data_name + namDN + '/data.csv' 
            df_2 = pd.read_csv(full_path, encoding='utf-8', sep=',')
            df_2 = pd.concat([df_1, df_2])
            df_2['name'] = namDN
            df_2['date'] = data
            new_df = pd.concat([new_df, df_2])
            
```
